```python
import easyocr
import cv2
import re
import matplotlib.pyplot as plt

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'], gpu=False)  # Set gpu=True if you have GPU support

# Preprocess the image (Optional)
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.GaussianBlur(img, (5, 5), 0)  # Apply blur
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    return gray

# Extract digits using EasyOCR
def extract_digits(image_path):
    img = preprocess_image(image_path)
    results = reader.readtext(img, detail=0)  # Extract text without bounding box details

    # Filter only digits using regex
    digits_only = ''.join(re.findall(r'\d+', ''.join(results)))
    return digits_only

# Test the function on your images
image_path_1 = "VW_DTN01_L_00007_101000000000000000002038853600000.front.JPG"
image_path_2 = "VW_DTN01_A_00002_101000000000000000002051033000000.front.JPG"

print("Digits in Image 1:", extract_digits(image_path_1))
print("Digits in Image 2:", extract_digits(image_path_2))

# Optional: Display the image
def display_image(image_path):
    img = cv2.imread(image_path)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

display_image(image_path_1)
display_image(image_path_2)
```
